```python
import requests
from bs4 import BeautifulSoup

link = "https://en.wikipedia.org/wiki/Python_(programming_language)"
page = requests.get(link)
soup = BeautifulSoup(page.content, 'html.parser')


def get_wikipedia_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        return None

def extract_wikipedia_content(soup):
    article_dict = {}
    main_title = soup.find("span", {"class": "mw-page-title-main"}).get_text()
    article_dict[main_title] = []
    headings = soup.find_all(class_="mw-headline")
    paragraphs = soup.find_all("p")
    current_heading = None

    for element in headings + paragraphs:
        if element in headings:
            current_heading = element.text
            article_dict[current_heading] = []
        elif current_heading:
            article_dict[current_heading].append(element.text)

    return article_dict

def find_redirect_links(url):
    page_content = get_wikipedia_page(url)
    if page_content is None:
        return []

    soup = BeautifulSoup(page_content, 'html.parser')
    redirect_links = []

    for anchor in soup.find_all('a', href=True):
        href = anchor['href']
        if href.startswith('/wiki/'):
            redirect_links.append(href)

    return redirect_links

def main(url):
    page_content = get_wikipedia_page(url)
    if page_content is None:
        print("Failed to fetch the Wikipedia page.")
        return

    soup = BeautifulSoup(page_content, 'html.parser')
    extracted_content = extract_wikipedia_content(soup)

    if extracted_content:
        for heading, paragraphs in extracted_content.items():
            print(heading)
            for paragraph in paragraphs:
                print(paragraph)

    # Find and print links that redirect to other Wikipedia pages
    redirect_links = find_redirect_links(url)
    
    if redirect_links:
        print("\nLinks that redirect to other Wikipedia pages:")
        for link in redirect_links:
            print(f"https://en.wikipedia.org{link}")

if __name__ == "__main__":
    link = "https://en.wikipedia.org/wiki/Python_(programming_language)"
    main(link)
```
