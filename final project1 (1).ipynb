```python
import tkinter as tk
from tkinter import simpledialog
from random import randint

choix_joueur = None
choix_pc = None

def jeux():
    global choix_joueur, choix_pc
    choix_joueur = simpledialog.askstring("Pierre, feuille, ciseaux", "Choisissez : pierre, feuille ou ciseaux")

    if choix_joueur:
        player_choice_label.config(text="Le joueur a choisi : " + choix_joueur)
        choix_pc = computer()
        computer_choice_label.config(text="L'ordinateur a choisi : " + choix_pc)
        result = determine_winner(choix_joueur, choix_pc)
        result_label.config(text=result)

    else:
        player_choice_label.config(text="")
        computer_choice_label.config(text="")
        result_label.config(text="Le joueur n'a pas choisi.")

def computer():
    comp_pick = randint(1, 3)
    if comp_pick == 1:
        return "pierre"
    elif comp_pick == 2:
        return "feuille"
    else:
        return "ciseaux"

def determine_winner(player, computer):
    if player == computer:
        return "C'est une égalité !"
    elif (player == "pierre" and computer == "ciseaux") or (player == "feuille" and computer == "pierre") or (player == "ciseaux" and computer == "feuille"):
        return "Le joueur gagne !"
    else:
        return "L'ordinateur gagne !"

def Reset():
    global choix_joueur, choix_pc
    choix_joueur = None
    choix_pc = None
    player_choice_label.config(text="")
    computer_choice_label.config(text="")
    result_label.config(text="Le jeu a été réinitialisé.")

def exit_game():
    window.destroy()

window = tk.Tk()
window.title("Jeu de Pierre, Feuille, Ciseaux")
width = 800
height = 600
window.geometry(f"{width}x{height}")
background_color = "blue"
window.configure(bg=background_color)
title_label = tk.Label(window, text="Habib's game", font=("New York", 24), fg="white", bg=background_color)
title_label.pack(pady=20)

play_button = tk.Button(window, text="Jouer", command=jeux)
play_button.pack()

reset_button = tk.Button(window, text="Réinitialiser", command=Reset)
reset_button.pack()

exit_button = tk.Button(window, text="Exit", command=exit_game)
exit_button.pack()

player_choice_label = tk.Label(window, text="", font=("New York", 18), fg="white", bg=background_color)
player_choice_label.pack()

computer_choice_label = tk.Label(window, text="", font=("New York", 18), fg="white", bg=background_color)
computer_choice_label.pack()

result_label = tk.Label(window, text="", font=("New York", 18), fg="white", bg=background_color)
result_label.pack()

window.mainloop()
```
